```python
import matplotlib.pyplot as plt
import numpy as np


metrics = ['Precision', 'Recall', 'F1-score', 'MCC']

lr = [0.8394, 0.7815, 0.8094, 0.7902]
svm = [0.8835, 0.7786, 0.8277, 0.8124]
knn = [0.8819, 0.8431, 0.8621, 0.8478]
gcn = [0.8941, 0.7551, 0.8188, 0.8044]
graphsage = [0.9214, 0.8255, 0.8708, 0.8593]
gat = [0.8977, 0.7463, 0.8151, 0.8012]
proposed = [0.9423, 0.8622, 0.9005, 0.8913]


models = ['LR', 'SVM', 'KNN', 'GCN', 'GAT', 'GraphSAGE', 'Proposed model']
values = [lr, svm, knn, gcn, gat, graphsage, proposed]

colors = ['#4A90E2', '#B0B0B0', '#9BC2E6', '#D9D9D9', '#A0A0A0', '#5B9BD5', '#4A90E2']
hatches = ['', '', '', '', '', '', '///']


x = np.arange(len(metrics))
width = 0.1

# 创建图形
plt.figure(figsize=(12, 6))

# 绘图
for i, (model, val) in enumerate(zip(models, values)):
    plt.bar(x + i * width, val, width, label=model, color=colors[i], hatch=hatches[i])

# 设置图表属性
plt.ylabel('Score')
plt.xticks(x + width * 3, metrics)
plt.ylim(0.7, 1.0)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.12), ncol=4)
plt.tight_layout()

# 保存高清图（可选）
# plt.savefig("./result/model_comparison_barplot.png", dpi=300, bbox_inches='tight')

# 显示图表
plt.show()
```
